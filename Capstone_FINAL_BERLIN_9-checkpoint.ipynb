## Introduction: Business Problem

Moving to a new city can elicit feelings of depression, anxiety, and loneliness. Foreigners moving to Berlin, in search of a therapist with whom they can discuss these issues, have historically had a hard time finding a therapist in a timely fashion. Compounding the general city-wide shortage of qualified psychotherapists is the added hardship of finding a therapist that is a native speaker of the language of their country of origin.

Focusing on the sizeable native English-speaking population of Berlin, I will use Foursquare API, Folium, and GeoPandas mapping software in order to determine the best place to open up a new English-speaking psychotherapeutic practice.

## Create a Table of Foreigners living in Berlin whose native tongue is English

As we are only interested in those people living in Berlin who would prefer to go to a therapist whose native langauge is English, we need to limit our database to those foreigners whose home country is an English speaking one. 

In order to do so, we will use Wikipedia's list of "English-speaking countries and territories." Given Britain's long history of colonialism, and the subsequent use, throughout the world, of English as a de facto lingua franca, this gives us a total of 131 countries or territories. 

This needs to be narrowed down. Let us instead restrict our database to those countries in which English functions not only as lingua franca between groups of people who do not share a native language or dialect, but as the primary natively spoken language. This new restrictive criteria yields a total of 17 countries. Much more workable!  

### We are only interested in these 17 countries, so let's remove all other nationalities from our database.  

```python
df_berlin = df[['Kontinente/Staaten','Irland','Vereinigtes Königreich','Kanada','Vereinigte Staaten / USA','Antigua und Barbuda','Bahamas','Barbados','Belize','Dominica','Grenada','Jamaika','St. Kitts und Nevis','St. Vincent und die Grenadinen','Trinidad und Tobago','Guyana','Australien','Neuseeland']]
df_berlin.head()
```

Let's remove all null values from our database, and then we are ready to map out the distribution of our native speakers of English.

```python
df_berlin_clean = df_berlin.dropna().reset_index(drop=True)
df_berlin_clean.head()
```

And finally, for clarity, let's change the title of the first column.

```python
df_berlin_clean.rename(columns = {'Kontinente/Staaten':'Neighborhood'}, inplace = True)
```

Let's also remove the unnecessary numbers appendant to our list of neighborhoods.

```python
df_berlin_clean["Neighborhood"] = df_berlin_clean['Neighborhood'].str.replace('\d+', '')
df_berlin_clean.head()
```

Let's turn all of our floats into integers to avoid problems down the line when we sum our totals. 

```python
#In order to do this, we create a new temporary database that includes only our numerical data.
new_df = df_berlin_clean[df_berlin_clean.columns[2:19]]

#Then, we transform our dashes to NaN.
new_df = new_df.apply (pd.to_numeric, errors='coerce')

#And, finally, replace all NaN with zeros. 
zeroed_df = new_df.replace(np.nan,0)
zeroed_df.head()
```

Finally, we re-join our numerical data with our 'Neighborhood' column.

```python
final_berlin = pd.concat([df_berlin_clean['Neighborhood'],zeroed_df], axis=1)
final_berlin.head()
```

## Berlin has 12 neighborhoods. Let's map them!

### First, use geopy to retrieve the latitude and longitude values of our neighborhoods in Berlin.

```python
geolocator = Nominatim(user_agent="Berlin_user")
final_berlin['Latitude'] = final_berlin['Neighborhood'].apply(geolocator.geocode).apply(lambda x: (x.latitude))
final_berlin['Longitude'] = final_berlin['Neighborhood'].apply(geolocator.geocode).apply(lambda x: (x.longitude))

final_berlin.head()
```

### We will use Foursquare API to locate where in Berlin psychotherapists are located. Foursquare does not have a separate category for psychotherapists, but all psychotherapists in Germany are licensed as Heilpraktiker, or Alternative Healers, which **is** a listed category.   

#### Category ID for 'Alternative Healers' is taken from Foursquare web site (https://developer.foursquare.com/docs/resources/categories)
52e81612bcbc57f1066b7a3c    is the code for the category 'Alternative Healer'

```python
address = 'Auguststrasse 10, Berlin, Germany'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

```python
radius = 100000
categoryId = '52e81612bcbc57f1066b7a3c'
url = 'https://api.foursquare.com/v2/venues/search?categoryId={}&intent=browse&client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(categoryId, CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
url;
```

```python
results = requests.get(url).json()
results;
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
psych = json_normalize(venues)
psych.head()
```

## Let's clean up this database.

```python
psych.columns = [column.split('.')[-1] for column in psych.columns]

new_psych = psych[psych.city == "Berlin"]
new_psych.head()
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=11)

# Add a red circle marker to represent 'Auguststrasse 10, which, for nostalgic reasons, I am 
# using as an approximate center of Berlin.' 
# Alternately, one can find the exact center of Berlin by calculating the mean of the centre of all features.
# x_map=gdf.centroid.x.mean()
# y_map=gdf.centroid.y.mean()
# folium.Map(location=[y_map, x_map], zoom_start=11)

folium.features.CircleMarker(
        [latitude, longitude],
        radius=10,
        color='red',
        popup="Middle of Berlin",
        fill = True,
        fill_color = 'red',
        fill_opacity = 0.6
    ).add_to(venues_map)

# add Psychotherapists as blue circle markers
for lat, lng, label in zip(new_psych.lat, new_psych.lng, new_psych.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

venues_map
```
